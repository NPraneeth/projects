```python
G = mg.read_csv('Golden_table.csv', ltable=A, rtable=B)
```

```python
IJ = mg.train_test_split(G, train_proportion=0.85)
```

```python
(len(I), len(J))
```

```python
dt = mg.DTMatcher(name='DecisionTree', random_state=0)
```

```python
svm = mg.SVMMatcher()
```

```python
rf = mg.RFMatcher(name='RF', random_state=0)
```

```python
nb = mg.NBMatcher(name='NB')
```

```python
lg = mg.LogRegMatcher(name='LogReg')
```

```python
ln = mg.LinRegMatcher(name='LinReg')
```

```python
(dt.name, svm.name)
```

```python
feat_table = mg.get_features_for_matching(A, B)
```

```python
feat_table
```

```python
mg._match_s
```

```python
mg._match_t
```

```python
mg._match_c['corres']
```

```python
feat_subset_iter1 = feat_table[3:6].append(feat_table[7:11]).append(feat_table[12:15].append(feat_table[37:45]))
```

```python
#feat_subset_iter3 = feat_table[7:15]
```

```python
#feat_subset_iter1 = feat_table[3:5,7:10,12:15]
```

```python
K = mg.extract_feature_vecs(I, feature_table=feat_subset_iter1, attrs_after='gold')
```

```python
K
```

```python
K.fillna(0, inplace=True)
```

```python
K
```

```python
result = mg.select_matcher([dt,rf,svm, nb, lg, ln], 
                           table=K,
                           exclude_attrs=['_id', 'ltable.ID', 'rtable.ID', 'gold'],
                           target_attr='gold', metric='precision', random_state=0)
```

```python
#result
```

```python
result['selected_matcher']
```

```python
result['cv_stats']
```

```python
#Debug the RF

# Split feature vectors to train and test
UV = mg.train_test_split(K, train_proportion=0.7, random_state=0)
U = UV['train']
V = UV['test']
```

```python
len(U)
```

```python
len(V)
```

```python
# Train rf using U
rf.fit(table=U, 
       exclude_attrs=['_id', 'ltable.id', 'rtable.id', 'gold'], 
       target_attr='gold')
```

```python
# Predict V using rf
P = rf.predict(table=V, exclude_attrs=['_id', 'ltable.id', 'rtable.id', 'gold'], 
              append=True, target_attr='predicted', inplace=False)
```

```python
P
```

```python
mg.vis_debug_rf(rf, U, V, 
        exclude_attrs=['_id', 'ltable.ID', 'rtable.ID', 'gold'],
        target_attr='gold')
```

```python
eval_result = mg.eval_matches(P, 'predicted', 'gold')
mg.print_eval_summary(eval_result)
```

```python
#Using RF as the best matcher
result = mg.cv_matcher_and_trigger(rf,[], table = K, 
                                   exclude_attrs=['_id', 'ltable.id', 'rtable.id', 'gold'],
                                  target_attr='gold')
result['cv_stats']
```

```python
#Adding triggers
```

```python
# Add trigger - target false positives: use title related feature
neg_trigger = mg.MatchTrigger()
neg_trigger.add_cond_rule('Title_Title_jac_qgm_3_qgm_3(ltuple, rtuple) < 0.85', feat_table)
neg_trigger.add_cond_status(True)
neg_trigger.add_action(0)
```

```python
# Do cross-validation for matcher + trigger using I (K)
result = mg.cv_matcher_and_trigger(rf, neg_trigger, table = K, 
                                   exclude_attrs=['_id', 'ltable.id', 'rtable.id', 'gold'],
                                  target_attr='gold')
```

```python
result['cv_stats']
```

```python
# Recall the cv for just the matcher (without trigger) was
result = mg.cv_matcher_and_trigger(rf, [], table = K, 
                                   exclude_attrs=['_id', 'ltable.id', 'rtable.id', 'gold'],
                                  target_attr='gold')
```

```python
result['cv_stats']
```

```python
# Now Z is rf (features: feat_subset_iter2)
#Not considering the negative trigger as it is not giving better result.
#It is lowering the recall and hence the accuracy.
```

```python
# Predict M 
N = rf.predict(table=M, exclude_attrs=['_id', 'ltable.ID', 'rtable.ID', 'gold'], 
              append=True, target_attr='predicted', inplace=False)
```

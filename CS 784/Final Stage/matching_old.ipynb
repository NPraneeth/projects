```python
G = mg.read_csv('Golden_table.csv', ltable=A, rtable=B)
```

```python
G
```

```python
IJ = mg.train_test_split(G, train_proportion=0.7)
I = IJ['train']
J = IJ['test']
```

```python
(len(I), len(J))
```

```python
dt = mg.DTMatcher(name='DecisionTree', random_state=0)
rf = mg.RFMatcher()
svm = mg.SVMMatcher()
nb = mg.NBMatcher(name='NB')
lg = mg.LogRegMatcher(name='LogReg')
ln = mg.LinRegMatcher(name='LinReg')
```

```python
feat_table = mg.get_features_for_matching(A, B)
```

```python
feat_table
```

```python
feat_subset_iter1 = feat_table[3:7].append(feat_table[7:12]).append(feat_table[12:15].append(feat_table[37:45]))
```

```python
feat_subset_iter1
```

```python
K = mg.extract_feature_vecs(I, feature_table=feat_subset_iter1, attrs_after='gold')
```

```python
K.fillna(0, inplace=True)
```

```python
result = mg.select_matcher([dt,rf,svm, nb, lg, ln], 
                           table=K,
                           exclude_attrs=['_id', 'ltable.ID', 'rtable.ID', 'gold'],
                           target_attr='gold', metric='precision', random_state=0)
```

```python
result['cv_stats']
```

```python
#Debug the RF

# Split feature vectors to train and test
UV = mg.train_test_split(K, train_proportion=0.5, random_state=0)
U = UV['train']
V = UV['test']
```

```python
# Train rf using U
rf.fit(table=U, 
       exclude_attrs=['_id', 'ltable.ID', 'rtable.ID', 'gold'], 
       target_attr='gold')
```

```python
# Predict V using rf
P = rf.predict(table=V, exclude_attrs=['_id', 'ltable.ID', 'rtable.ID', 'gold'], 
              append=True, target_attr='predicted', inplace=False)
```

```python
mg.vis_debug_rf(rf, 
                U, V, 
        exclude_attrs=['_id', 'ltable.ID', 'rtable.ID', 'gold'],
        target_attr='gold')
```

```python
eval_result = mg.eval_matches(P, 'predicted', 'gold')
mg.print_eval_summary(eval_result)
```

```python
finaltrain = rf.predict(table=K, exclude_attrs=['_id', 'ltable.ID', 'rtable.ID', 'gold'], 
              append=True, target_attr='predicted', inplace=False)
```

```python
eval_result = mg.eval_matches(finaltrain, 'predicted', 'gold')
mg.print_eval_summary(eval_result)
```

```python
#Adding triggers
```

```python
# Add trigger - target false positives: use title related feature
neg_trigger = mg.MatchTrigger()
#neg_trigger.add_cond_rule('Title_Title_jac_qgm_3_qgm_3(ltuple, rtuple) >0.5 && Publisher_Publisher_lev > 0.8 && Author_Author_lev > 0.8', feat_table)
#neg_trigger.add_cond_rule('Title_Title_jac_qgm_3_qgm_3(ltuple, rtuple) >0.5',feat_table)
neg_trigger.add_cond_rule('Publisher_Publisher_lev(ltuple, rtuple) > 0.8',feat_table) 
neg_trigger.add_cond_rule(['Title_Title_jac_qgm_3_qgm_3(ltuple, rtuple) >0.5', 
                           'Publisher_Publisher_lev(ltuple, rtuple) > 0.8'])
#neg_trigger.add_cond_rule('Author_Author_lev > 0.8', feat_table)
neg_trigger.add_cond_status(True)
neg_trigger.add_action(0)
```

```python
# Do cross-validation for matcher + trigger using I (K)
result = mg.cv_matcher_and_trigger(rf, neg_trigger, table = M, 
                                   exclude_attrs=['_id', 'ltable.ID', 'rtable.ID', 'gold'],
                                  target_attr='gold')
```
